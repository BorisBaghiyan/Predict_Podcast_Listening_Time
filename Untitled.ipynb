```python
df_test = pd.read_csv('test.csv')
```

```python
df_train = df_train.drop('id',axis = 1)
df_test = df_test.drop('id',axis = 1)
```

```python
df_train = df_train.drop('Episode_Title',axis = 1)
df_test = df_test.drop('Episode_Title',axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df_train['Podcast_Name'] = label.fit_transform(df_train['Podcast_Name'])
df_test['Podcast_Name'] = label.fit_transform(df_test['Podcast_Name'])
```

```python
df_train["Episode_Length_minutes"] = df_train["Episode_Length_minutes"].fillna(df_train['Episode_Length_minutes'].median())
```

```python
df_test["Episode_Length_minutes"] = df_test["Episode_Length_minutes"].fillna(df_test['Episode_Length_minutes'].median())
```

```python
df_train['Genre'] = label.fit_transform(df_train['Genre'])
df_test['Genre'] = label.fit_transform(df_test['Genre'])
```

```python
df_train['Publication_Day'] = label.fit_transform(df_train['Publication_Day'])
df_test['Publication_Day'] = label.fit_transform(df_test['Publication_Day'])
```

```python
df_train['Publication_Time'] = label.fit_transform(df_train['Publication_Time'])
df_test['Publication_Time'] = label.fit_transform(df_test['Publication_Time'])
```

```python
df_train["Guest_Popularity_percentage"] = df_train["Guest_Popularity_percentage"].fillna(df_train['Guest_Popularity_percentage'].median())
df_test["Guest_Popularity_percentage"] = df_test["Guest_Popularity_percentage"].fillna(df_test['Guest_Popularity_percentage'].median())
```

```python
df_train["Number_of_Ads"] = df_train["Number_of_Ads"].fillna(df_train['Number_of_Ads'].median())
df_test["Number_of_Ads"] = df_test["Number_of_Ads"].fillna(df_test['Number_of_Ads'].median())
```

```python
df_train['Episode_Sentiment'] = label.fit_transform(df_train['Episode_Sentiment'])
df_test['Episode_Sentiment'] = label.fit_transform(df_test['Episode_Sentiment'])
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
X_train = df_train.drop('Listening_Time_minutes',axis = 1)
y_train = df_train["Listening_Time_minutes"]
X_test = df_test.copy()
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
train_pred = model.predict(X_train)
train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
```
